```python
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns

df = pd.DataFrame(pd.read_excel("ZF_Elec_Responses.xlsx"))
df
```

```python
print(df.dtypes), df.nunique()
```

```python
df.drop(columns = {'Phone Number', 'Email', 'Facebook URL', 'LinkedIn URL (optional)', 'Any questions?'}, inplace = True)

df.head(2)
```

```python
df['Academic Year'].unique(), df['Do you have any volunteering experience?'].unique(), df['How did you hear about us?'].unique()
```

```python
plt.bar(df['Do you have any volunteering experience?'].value_counts().index, df['Do you have any volunteering experience?'].value_counts().values, label="Do you have any volunteering experience?")
plt.xlabel("Answers")
plt.legend()
plt.ylabel("Numbers")
plt.show()
```

```python
df.groupby(['Do you have any volunteering experience?'])['Academic Year'].value_counts().unstack().plot(kind = 'bar',
                                                                           title = 'Relation between Academic Year and Volunteer Experience', 
                                                                           figsize = (8, 8));
```

```python
plt.bar(df['Academic Year'].value_counts().index, df['Academic Year'].value_counts().values, label="Academic Year")
plt.xlabel("Academic Year")
plt.legend()
plt.ylabel("Values")
plt.show()
```

```python
facebook_rows = df[df['How did you hear about us?'].str.contains('Facebook')]
Friends_rows = df[df['How did you hear about us?'].str.contains('Friends')]
Faculty_colleagues_rows = df[df['How did you hear about us?'].str.contains('Faculty colleagues')]
Instagram_rows = df[df['How did you hear about us?'].str.contains('Instagram')]


print(facebook_rows.shape[0])
print(Friends_rows.shape[0])
print(Faculty_colleagues_rows.shape[0])
print(Instagram_rows.shape[0])
```

```python
plt.bar("facebook", facebook_rows.shape[0], label="facebook")
plt.bar("Friends", Friends_rows.shape[0], label="Friends")
plt.bar("Faculty_colleagues", Faculty_colleagues_rows.shape[0], label="Faculty_colleagues")
plt.bar("Instagram", Instagram_rows.shape[0], label="Instagram")

plt.xlabel("about us")
plt.legend()
plt.ylabel("Values")
plt.show()
```

```python
Electrical_rows = df[df['Faculty - Department'].str.contains('Electrical|electric|Electricity|electrical|الكهربائية')]
Mechatronics_rows = df[df['Faculty - Department'].str.contains('Mechatronics|ميكاترونكس')]
prep_rows = df[df['Faculty - Department'].str.contains('prep|اعدادي')]
mechanical_rows = df[df['Faculty - Department'].str.contains('mechanical|ميكانيكا')]
other_dep = df.shape[0] - (Electrical_rows.shape[0] + Mechatronics_rows.shape[0] + prep_rows.shape[0] + mechanical_rows.shape[0])

print(Electrical_rows.shape[0])
print(Mechatronics_rows.shape[0])
print(prep_rows.shape[0])
print(mechanical_rows.shape[0])
print(other_dep)
```

```python
plt.bar("Electrical", Electrical_rows.shape[0], label="Electrical")
plt.bar("Mechatronics", Mechatronics_rows.shape[0], label="Mechatronics")
plt.bar("prep", prep_rows.shape[0], label="prep")
plt.bar("mechanical", mechanical_rows.shape[0], label="mechanical")
plt.bar("other_dep", other_dep, label="other_dep")

plt.xlabel("Departments")
plt.legend()
plt.ylabel("Numbers")
plt.show()
```
